```python
temp = ['Bowlegs', 'Knock-knee', 'Normal Knee']
p = "F:\\2022-2023\\SEM 1\\CV\\CV_FINAL\\Dataset\\"
for i in temp:
    print("Class = ", i)
    for j in (os.listdir(p + i)):
        org = cv2.imread(p + i + '\\' + j)
        gray = cv2.cvtColor(org, cv2.COLOR_BGR2GRAY)
        pr = cv2.resize(gray, (128, 128))
        path = "F:\\2022-2023\\SEM 1\\CV\\Dataset\\"+i
        cv2.imwrite(os.path.join(path, j), pr)
```

```python
for i in temp:
    path = "F:\\2022-2023\\SEM 1\\CV\\Dataset\\"+i
    for j in os.listdir(path):
        print(j)
        img = cv2.imread(path + "\\" +j)
        gray = skimage.color.rgb2gray(img)
        print(gray.shape)
```

```python
for i in temp:
    p = "F:\\2022-2023\\SEM 1\\CV\\Dataset\\"
    data_gen = ImageDataGenerator(
        horizontal_flip=True,
        vertical_flip=True,
        rotation_range=20,
        fill_mode='nearest',
        zoom_range=0.15,
        shear_range=0.15,
    )
    for j in os.listdir(p+i):
        print(p+i+"\\"+ j)
        pic = load_img(p+i+"\\"+ j)
        pic_array = img_to_array(pic)
        pic_array = pic_array.reshape((1,) + pic_array.shape)
        count = 0
        for batch in data_gen.flow( pic_array, batch_size = 1, save_to_dir = r'F:\2022-2023\SEM 1\CV\Dataset_1\Augmented\\' + i,  save_prefix=i, save_format='png'):
            count += 1
            if count >= 10:
                break

path1 = r"F:\2022-2023\SEM 1\CV\Dataset\\"
path2 = "F:\\2022-2023\\SEM 1\\CV\\Dataset\\Knock-knee\\"
path3 = "F:\\2022-2023\\SEM 1\\CV\\Dataset\\Normal Knee\\"
# pp1 = data_gen.flow_from_directory(directory=path1, save_to_dir=r"F:\2022-2023\SEM 1\CV\Dataset_1\Augmented\\", save_format="png")
# for i in pp1:
#     print(i)
```
